```python
print("==================================================")
print(" 🧠 BIENVENIDO AL ASISTENTE DE CARPINTERÍA 'IA' 🧡 ")
print("==================================================")
print("Este sistema simula el pensamiento de un maestro carpintero.")
print("Te ayudaré a diseñar el mueble ideal para tu espacio.")

# Variable para mantener el programa funcionando
ejecutando = "si"

while ejecutando == "si":
    print("\n--- MENÚ PRINCIPAL ---")
    print("1. Consultar al Asistente de IA (Diseño de Muebles)")
    print("2. Ver tipos de madera disponibles")
    print("3. Salir del programa")

    opcion = input("\nSelecciona una opción (1, 2 o 3): ")

    # -----------------------------------------------------------------
    # OPCIÓN 1: EL MOTOR DE "INTELIGENCIA ARTIFICIAL" (SISTEMA EXPERTO)
    # -----------------------------------------------------------------
    if opcion == "1":
        print("\n--- INICIANDO DIAGNÓSTICO DE IA ---")
        print("Por favor, responde las siguientes preguntas para calcular tu mueble ideal.")

        # Pregunta 1: Espacio
        print("\n¿Dónde vas a colocar el mueble?")
        print("A. En la sala de estar")
        print("B. En el dormitorio")
        lugar = input("Tu respuesta (A o B): ")

        # Pregunta 2: Presupuesto
        print("\n¿Cuál es tu presupuesto estimado?")
        print("A. Ajustado / Económico")
        print("B. Alto / Premium")
        presupuesto = input("Tu respuesta (A o B): ")

        print("\n[IA Procesando datos...] 🧠🔨")

        # Árbol de decisión interactivo (Aquí ocurre la "magia" de la IA)
        if lugar == "A" or lugar == "a":
            if presupuesto == "A" or presupuesto == "a":
                print("\n💡 RESULTADO DE LA IA:")
                print("Te recomiendo una -> Mesa de Centro de Pino Económica.")
                print("Ideal para salas pequeñas, ligera y muy accesible.")
            elif presupuesto == "B" or presupuesto == "b":
                print("\n💡 RESULTADO DE LA IA:")
                print("Te recomiendo un -> Centro de Entretenimiento de Madera de Cedro.")
                print("Un mueble elegante, resistente y con un aroma increíble para tus visitas.")
            else:
                print("\n❌ Opción de presupuesto no válida. Inténtalo de nuevo.")

        elif lugar == "B" or lugar == "b":
            if presupuesto == "A" or presupuesto == "a":
                print("\n💡 RESULTADO DE LA IA:")
                print("Te recomiendo un -> Buró de Noche Minimalista (MDF Estilizado).")
                print("Práctico, moderno y no gasta mucho espacio ni dinero.")
            elif presupuesto == "B" or presupuesto == "b":
                print("\n💡 RESULTADO DE LA IA:")
                print("Te recomiendo un -> Ropero / Clóset Amplio de Madera de Roble.")
                print("Para toda la vida. Máxima durabilidad y elegancia para tu habitación.")
            else:
                print("\n❌ Opción de presupuesto no válida. Inténtalo de nuevo.")

        else:
            print("\n❌ Ubicación no reconocida. El carpintero se confundió.")

    # -----------------------------------------------------------------
    # OPCIÓN 2: CONSULTA DE MATERIALES
    # -----------------------------------------------------------------
    elif opcion == "2":
        print("\n--- CATÁLOGO DE MATERIALES RECOMENDADOS POR LA IA ---")
        print("- Pino: Económico y fácil de trabajar. Ideal para principiantes.")
        print("- MDF: Planchas de fibra de madera, baratas y perfectas para pintar.")
        print("- Cedro: Color rojizo, resiste la humedad y las polillas. Huele genial.")
        print("- Roble: Ultra resistente, pesado y con hermosas vetas oscuras.")

    # -----------------------------------------------------------------
    # OPCIÓN 3: SALIR
    # -----------------------------------------------------------------
    elif opcion == "3":
        print("\n¡Gracias por usar el Asistente Carpintero IA! Hasta luego. 🧡🧠")
        ejecutando = "no" # Rompe el ciclo while de forma limpia

    # CONTROL DE ERRORES DEL MENÚ PRINCIPAL
    else:
        print("\n❌ Opción inválida. Por favor, elige 1, 2 o 3.")

    # Preguntar si desea continuar si no ha elegido salir
    if opcion != "3":
        print("\n--------------------------------------------------")
        ejecutando = input("¿Deseas volver al menú principal? (si/no): ")
        if ejecutando != "si" and ejecutando != "SI" and ejecutando != "Si":
            print("\n¡Gracias por usar el software! ¡Adiós!")
            ejecutando = "no"
print("Programa elaborado por Keyla Cigarroa NL = 12")
```
